```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#plt.plot(x,y , label='2x', color='red' , linewidth=2, marker='.', markersize=10, linestyle='--', markeredgecolor='blue')
# Resize graph
plt.figure(figsize=(5,3), dpi=150)
#Use Shorthand notation
plt.plot(x,y, 'b^--' , label='2x')

#Line Number Two
x2 = np.arange(0,4,0.5)
print(x2)
plt.plot(x2[:6], x2[:6]**2,'r', label='X^2')
plt.plot(x2[5:],x2[5:]**2, 'r--')
plt.title('Our First Graph!', fontdict={'fontname': 'Comic sans MS', 'fontsize':20})

plt.xlabel('X Axis!')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.savefig('mygraph.png', dpi=200)
plt.show()
```

```python
labels = ['A','B','C']
values = [1,3,5]

bars =plt.bar(labels, values)

patters = ['/','o','*']

for bar in bars:
    bar.set_hatch(patters.pop(0))
# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,9))
plt.show()
```
